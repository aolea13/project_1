```python
file_path = "data/zhvi_data.csv"
home_values_df=pd.read_csv(file_path)

home_values_df.head()
```

```python
fl_filter=home_values_df['State']=="FL"
fl_df=home_values_df[fl_filter]
display(fl_df)
```
