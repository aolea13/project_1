```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "data/Metro_invt_fs_uc_sfrcondo_smoothed_month.csv"

sales_inventory_df=pd.read_csv(file_path)

sales_inventory_df.head()
```

```python
data_columns= ['2018-01-31', '2018-02-28', '2018-03-31', '2018-04-30', '2018-05-31',
       '2018-06-30', '2018-07-31', '2018-08-31', '2018-09-30', '2018-10-31',
       '2018-11-30', '2018-12-31', '2019-01-31', '2019-02-28', '2019-03-31',
       '2019-04-30', '2019-05-31', '2019-06-30', '2019-07-31', '2019-08-31',
       '2019-09-30', '2019-10-31', '2019-11-30', '2019-12-31', '2020-01-31',
       '2020-02-29', '2020-03-31', '2020-04-30', '2020-05-31', '2020-06-30',
       '2020-07-31', '2020-08-31', '2020-09-30', '2020-10-31', '2020-11-30',
       '2020-12-31']
```

```python
# Create a line plot showing For Sale inventory for Riverside Metro Area
riverside_filter=sales_inventory_df['RegionName']=="Riverside, CA"

riverside_inventory_over_time=sales_inventory_df.loc[riverside_filter, data_columns]

plt.figure(figsize=(12,8))
plt.plot(data_columns,riverside_inventory_over_time.iloc[0], marker="o")
plt.xlabel("Date", fontsize=10)
plt.ylabel("For Sale Inventory")
plt.title("Riverside Regional Sales Inventory 2018-2020")
plt.xticks(fontsize=8,rotation=90)

plt.show()
```
